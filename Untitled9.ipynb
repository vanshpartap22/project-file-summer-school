```python
# Flipkart Sales Analysis Project
# Author: [Vansh Partap Singh]
# Internship Project IIT Jammu Summer School 2025

import pandas as pd
import matplotlib.pyplot as plt
import random
import datetime

# ------------------------------
# Step 1: Generate Sample Dataset
# ------------------------------
def generate_sales_data(filename="flipkart_sales.csv", rows=100):
    categories = ["Electronics", "Fashion", "Home", "Beauty", "Books", "Toys"]
    products = {
        "Electronics": ["Smartphone", "Laptop", "Headphones", "Smartwatch"],
        "Fashion": ["Shirt", "Jeans", "Shoes", "Jacket"],
        "Home": ["Mixer", "Vacuum Cleaner", "Curtains", "Furniture"],
        "Beauty": ["Lipstick", "Perfume", "Face Cream", "Shampoo"],
        "Books": ["Novel", "Textbook", "Comics", "Biography"],
        "Toys": ["Action Figure", "Board Game", "Puzzle", "Doll"]
    }

    data = []
    start_date = datetime.datetime(2024, 1, 1)

    for i in range(rows):
        category = random.choice(categories)
        product = random.choice(products[category])
        price = random.randint(500, 80000)
        quantity = random.randint(1, 5)
        total = price * quantity
        date = start_date + datetime.timedelta(days=random.randint(0, 365))

        data.append([date, category, product, price, quantity, total])

    df = pd.DataFrame(data, columns=["Date", "Category", "Product", "Price", "Quantity", "Total_Sales"])
    df.to_csv(filename, index=False)
    print(f"Sample sales data generated in {filename}")

generate_sales_data()

# ------------------------------
# Step 2: Load and Explore Data
# ------------------------------
df = pd.read_csv("flipkart_sales.csv", parse_dates=["Date"])
print("First 5 rows of dataset:")
print(df.head())

# ------------------------------
# Step 3: Sales Analysis
# ------------------------------

# Total revenue
total_revenue = df["Total_Sales"].sum()
print("\nTotal Revenue: â‚¹", total_revenue)

# Sales by Category
sales_by_category = df.groupby("Category")["Total_Sales"].sum().sort_values(ascending=False)
print("\nSales by Category:\n", sales_by_category)

# Top 5 Products
top_products = df.groupby("Product")["Total_Sales"].sum().sort_values(ascending=False).head(5)
print("\nTop 5 Best-Selling Products:\n", top_products)

# Monthly Sales Trend
df["Month"] = df["Date"].dt.to_period("M")
monthly_sales = df.groupby("Month")["Total_Sales"].sum()
print("\nMonthly Sales Trend:\n", monthly_sales)

# ------------------------------
# Step 4: Visualization
# ------------------------------

# Sales by Category
plt.figure(figsize=(8,5))
sales_by_category.plot(kind="bar")
plt.title("Sales by Category")
plt.ylabel("Revenue (â‚¹)")
plt.show()

# Top 5 Products
plt.figure(figsize=(8,5))
top_products.plot(kind="pie", color="orange")
plt.title("Top 5 Best-Selling Products")
plt.ylabel("Revenue (â‚¹)")
plt.show()

# Monthly Sales Trend
plt.figure(figsize=(10,6))
monthly_sales.plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")
plt.ylabel("Revenue (â‚¹)")
plt.xlabel("Month")
plt.show()
```
